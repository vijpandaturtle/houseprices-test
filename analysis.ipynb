```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline 

pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('houseprices.csv')

print(data.shape)
```

```python
data.head()
```

```python
#Variables with missing values 
vars_na = [var for var in data.columns if data[var].isnull().sum()>1]
#percentage of missing values for each variable
for var in vars_na:
    print(var, np.round(data[var].isnull().mean(), 2), '% missing values')
```

## Importance of features with missing values

```python
def analyse_na_value(df, var):
    df = df.copy()
    df[var] = np.where(df[var].isnull(), 1, 0)
    #mean saleprice with missing info 
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.show()
    
for var in vars_na:
    analyse_na_value(data, var)
```

```python
num_vars = [var for var in data.columns if data[var].dtypes != 'O']
print('Number of numerical variables', len(num_vars))
data[num_vars].head()
```

```python
print('Number of House ID labels:', len(data.Id.unique()))
print('Number of houses in the dataset', len(data))
```

```python
#Temporal Variables
#List of variables containing year information 
year_vars = [var for var in num_vars if 'Yr' in var or 'Year' in var]
year_vars
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel('Median Price sold')
```

## Importance of Year(Temporal Values)

```python
#Analyzing relationships between year values in detail
def analyse_year_vars(df, var):
    df = df.copy()
    #diff between year variable and year the house was sold 
    df[var] = df['YrSold'] - df[var]
    
    plt.scatter(df[var], df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
    
for var in year_vars:
    if var != 'YrSold':
        analyse_year_vars(data, var)
```

## Importance of discrete variables

```python
#discrete variables 
discrete_vars = [var for var in num_vars if len(data[var].unique()) < 20 and var not in year_vars+['Id']]
print('Number of discrete variables :', len(discrete_vars))
```

```python
data[discrete_vars].head()
```

```python
def analyse_discrete(df, var):
    df = df.copy()
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(data, var)
```

## Importance of continuous variables

```python
#List of continuous vars extracted from num_vars 
cont_vars = [var for var in num_vars if var not in discrete_vars+year_vars+['Id']]
print('Number of continuous variables', len(cont_vars))
```

```python
data[cont_vars].head()
```

```python
def analyse_cont_vars(df, var):
    df = df.copy()
    df[var].hist(bins=20)
    plt.ylabel('Number of houses')
    plt.xlabel(var)
    plt.title(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_cont_vars(data, var)
```

## Change variable distribution from skewed to gaussian

Using log transformation

```python
def analyse_transformed_continuous(df, var):
    df = df.copy()
    
    #Log cannot take neg values
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df[var].hist(bins=20)
        plt.ylabel('Number of houses')
        plt.xlabel(var)
        plt.title(var)
        plt.show()
        
for var in cont_vars:
    analyse_transformed_continuous(data, var)
```

```python
#analyse transformed values 
def transform_analyse_continuous(df, var):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df['SalePrice'] = np.log(df['SalePrice'])
        plt.scatter(df[var], df['SalePrice'])
        plt.ylabel('SalePrice')
        plt.xlabel(var)
        plt.show()
        
for var in cont_vars:
    if var != 'SalePrice':
        transform_analyse_continuous(data, var)
```

## Outliers 

```python
#boxplots to visualize outliers in continuous variables
def find_outliers(df, var):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()
        
for var in cont_vars:
    find_outliers(data, var)
```

## Categorical Variables

```python
##categorical vars
cat_vars = [var for var in data.columns if data[var].dtypes == 'O']
print('Number of categorical vars :', len(cat_vars))
```

```python
data[cat_vars].head()
```

```python
#Number of labels per attribute 
for var in cat_vars:
    print(var, len(data[var].unique()), 'categories')
```

```python
#finding rare labels 
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp<rare_perc]

for var in cat_vars:
    print(analyse_rare_labels(data, var, 0.01))
    print()
```
